```python
def cut(A,y):
    C_0 = np.where(y==0)[0] # create cluster C_0
    C_1 = np.where(y==1)[0] # create cluster C_1

    cut_term = 0 #initial cut term is 0
    
    #sum up the entries A[i,j] for each pair (i,j) in clusters  C_0, C_1
    for i in C_0:
        for j in C_1:
            cut_term += A[i,j]
    return cut_term
```

```python
def vols(A,y):
    C_0 = np.where(y==0)[0] # create cluster C_0
    C_1 = np.where(y==1)[0] # create cluster C_1

    #returns the volume of C_0 and C_1 as tuple.
    #note that row-wise addition and column-wise addition will give the same result.
    return (A[C_0,:].sum(axis = 0).sum(),A[C_1,:].sum(axis = 0).sum())


```

```python
vols(A,y)
```

```python
def normcut(A,y):
    #find the volume of each cluster
    v_0,v_1 = vols(A,y)

    #find the cut_term 
    cut_val = cut(A,y)
    
    return cut_val * (1/v_0 + 1/v_1)
```

```python
def transform(A,y):

    #find the volume of each cluster
    v_0,v_1 = vols(A,y)

    #compute the appropriate z
    z = 1/v_0 - (1/v_0 + 1/v_1)*y

    return z
```

```python
D = np.diag(np.sum(A,axis = 1))
z = transform(A,y)

part_b = normcut(A,y)
part_c = (z@(D-A)@z)/(z@D@z)

np.isclose(part_b,part_c)
```

```python
np.isclose(z.T@D@np.ones(n),0)
```

```python
def orth(u, v):
    return (u @ v) / (v @ v) * v

e = np.ones(n) 

d = D @ e

def orth_obj(z):
    z_o = z - orth(z, d)
    return (z_o @ (D - A) @ z_o)/(z_o @ D @ z_o)
```

```python
from scipy.optimize import minimize

z_ = minimize(orth_obj, np.ones(n))
```

```python
z_min = z_.x
```

```python
plt.scatter(X[:,0],X[:,1], c = z_min<0)
```

```python
L = np.linalg.inv(D)@(D-A)
L
```

```python
#find the eigenvalues and eigenvectors of L
w,v = np.linalg.eig(L)

#find the eigenvector corresponding to its second-smallest eigenvalue w.
z_eig = v[:,np.argsort(w)][:,1]
z_eig
```

```python
plt.scatter(X[:,0],X[:,1], c = z_eig<0)
```
